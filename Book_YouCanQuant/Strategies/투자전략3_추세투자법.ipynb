## 투자전략 3 추세투자법

### Strategy:
3개월 이평선 상승반전 --> 지수 매수<br>
3개월 이평선 하락반전 --> 지수 매도(전액 현금보유)<br>
[스타일: 자산배분 + 모멘텀]

### 절대적 모멘텀, "The trend is your friend"

'상승장에서 사고 하락장에서 빠지는' 모멘텀(**Momentum**) 전략 혹은 추세추종(**Trend following**)

<br>
*"주가가 많이 떨어졌다고 더 떨어질 수 없는 것이 아니고,<br>
 주가가 엄청나게 비싸도 더 오르지 못하는 것이 아니다.<br>
 주가가 오르면 사람들은 더 오를 것 같아서 더 사들인다."*<br>
-- Joseph de la Vega, 네덜란드 무역상 (**1688年** 출간한 세계최초 주식투자서 'Confusion de Confusiones')

**18世記** 혼마 무네히사(本間宗久)도 쌀 선물 거래에 추세추종 기법 사용. (저서: 本間宗久翁秘録)

*"Cut short your losses and let your profits run on."*<br>
-- David Ricardo (**200年前**), 경제학의 'comparative advantage' 창시자

<br>
절대적 모멘텀: 장이 좋을 때 투자<br>
상대적 모멘텀: 시장과 상관없이 수익이 좋은 종목에 투자

'상승하는 주식'과 '비싼 주식'은 다른 개념<br>
상승했더라도 적정가보다 쌀 수 있기에

<br>
상승장인 하락장인지 판단하는 것은 투자자 자신이 기준을 설정해야 함<br>
- 신진오: 3개월 이평선
- 안토나치: 주가지수 6개월 or 12개월 수익률이 예금금리보다 높으면 상승장, 낮으면 하락장

--> 하나의 정답이 없다.

<br>
주식시장은 몇개월 또는 1~2년간 천천히 떨어지다가 어느 시점에 대폭락하는 경우가 대부분

대부분의 모멘텀 전략은 횡보장에서는 위력을 발회하지 못하나,<br>
상승장에서 벌고 하락장에서 손실을 피함 --> 장기적으로 수익률 높아지는 것

**중요**: 방어력이 뛰어나서 하락장에서 덜 깨지는 것도 CAGR 증가에 매우 큰 도움이 된다는 것

```python
from quantopian.research import prices, symbols
import pandas as pd
```

```python
# Query historical pricing data
close_price = prices(
    assets=symbols('SPY'),
    start='2002-01-01',
    end=  '2018-02-16',
)
```

```python
WINDOW = 60

# Compute moving averages
sma1 = close_price.rolling(WINDOW).mean()

# Form a DataFrame
df = pd.DataFrame({
    'SPY'  : close_price,
    'SMA1' : sma1,
})

# Plot
df[['SPY', 'SMA1']].plot(
    title='SPY close_price & SMA'
);
```
