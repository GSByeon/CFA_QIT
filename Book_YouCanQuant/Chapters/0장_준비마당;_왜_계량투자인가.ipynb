### 02 핵심 개념 장착하기

<br>대형주: 시총 상위 200위<br>
소형주: 시총 하위 20%

```python
# CAGR (Compound Annual Growth Rate)
initial_asset = 10000.
final_asset = 19500.
years = 3.

CAGR = (final_asset/initial_asset)**(1/years) - 1

print('CAGR = {}%'.format(round(CAGR*100, 2)))
```

**MDD** (Maximum Drawdown)

투자자의 '최대 고통'을 측정<br>
필자는 MDD를 줄이기 위해서는 수익의 일부를 양보해도 된다고 생각함

**방어형** 전략: MDD 축소에 초점<br>
**공격형** 전략: CAGR 극대화가 목표 (대신 MDD를 낮추는 별다른 안전망을 제공 않음)

<br>

알파 창출은 생각보다 어려워서, 펀드매니저 80~90%, 개인투자자 99% 이상이 실패

### 03 투자 수익을 좀먹는 생각의 함정

<br>우리의 두뇌는 합리적이지 않다.

카너먼(Daniel Kahneman) 교수는 우리 두뇌에 2가지 시스템이 존재한다고:<br>
직관체계(**Reflexive system**)와 추론체계(**Reflective system**)

추론체계는 의지가 없으면 작동하지 않는다. --> 노력과 주의가 필요

단순한 기계적 투자전략들이 수십 년 공부하고 경험을 가진 펀드매너지들보다 수익률이 훨씬 더 높다.

이 책의 전략들은 가격의 괴리(**Mispricing**)를 발생시키는 인간의 편향을 역이용해 알파를 창출한다.

투자는 일관성(**Consistency**)이 매우 중요

판사의 가석방 승인 통계: 식사 직후 65%, 식사 전 2시간 동안 떨어지다가, 식사 직전 0%<br>
--> 식사나 퇴근 시간 직전에는 결재 받지 말것!

과잉확신 편향(**Overconfidence bias**)<br>
한국 개인투자자 60% 이상이 1~2종목만 보유<br>
10개 종목 이하를 보유한 투자자 93%<br>
국내 주식투자자 500만명 중 책 읽고 공부하는 자는 5만명 미만

기준점 편향(**Anchoring bias**)

손실회피 편향(**Loss aversion bias**)<br>
같은 금액을 버는 기쁨보다 잃는 슬픔이 2.5~3배 정도 더 강하다고

사실 보유 주식을 매도할지 계속 보유할지의 결정에 매수 단가는 중요치 않다.<br>
합리적인 투자자는 투자전략에 적합하면 계속 보유하고, 그렇지 않으면 매도한다.

처분 효과(**Disposition effect**): 이익을 챙기는 데 급급한 행위

손실회피 편향 + 처분 효과 --> Peter Lynch: "꽃을 뽑아내고 잡초에 물 주는 행위"

확증 편향(**Confirmation bias**)<br>
자신의 신념과 일치하는 정보는 받아들이고 그렇지 않은 정보는 무시하는 것도 포함

주식투자가 위험한 것은, 어설픈 투자전략도 알파 달성 확률이 연단위 20~30%는 되고<br>
아무리 훌륭한 전략도 70~80% 밖에 안되기 때문이다.

인간은 통계 분석에 매우 약하다.

최신 편향(**Recency bias**): 최근 자료나 경험에 더 높은 비중을 둠

사람들은 그럴듯한 이야기, 스토리텔링에 매우 약하다.

### 04 두뇌 약점을 극복할 유일한 방법

카너먼 교수의 책 [생각에 관한 생각(Thinking, Fast and Slow)] 추천

주식을 매수하기 전에는 어느 정도 추론체계를 가동해 합리적이고 논리적 판단을 할 수 있다<br>
그러나 일단 주식을 사놓으면, 직관체계와 편향들이 투자자를 지배하게 된다.

전략은 매수하기 전, 결정해야 한다.<br>
또 어떤 조건이 되면 매도하는지 명시해야 한다.

모든 투자자에게 적합한 최고의 투자전략은 없다.<br>
성향, 멘털과 심리적 경향, 시장에 대한 철학, 경험, 능력, 자산, 수입 등 고려해야

<br>
계량투자의 함정: **과최적화**

**과최적화 피하는 방법:**<br>

(1) 3~4개 이상의 지표 활용 말라.<br>
'복잡한 전략+너무 아름다운 수익률'은 과최적화일 가능성 높음

(2) 외국 시장에서도 검증<br>

(3) 특히 전략이 공개된 후에도 알파가 유지되었는지 검증<br>

(4) 경제적(전략적) 논리가 합당해야<br>

(5) 지표를 조금 변형해 다시 백테스트<br>
PBR 0.5 이하 4개월 모멘텀 --> PBR 1 이하 6개월 모멘텀, PBR 0.3 이하 12개월 모멘텀<br>
또 한번은 PBR대신 PSR 같은 다른 지표를 써보라

<br>
자신이 구사하는 전략에 대한 믿음이 아주 강해야 한다<br>
잘 안 통하는 시기가 필히 오기 때문

워런 버핏이 갑부가 된 것은 전략을 계속 밀고 나갔기 때문임.<br>
